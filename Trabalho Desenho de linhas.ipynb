```python
#Neste código foi desenvolvido os dois códigos tanto para DDA quanto para Bresenham
import sys,pygame
from pygame import gfxdraw
pygame.init()
screen = pygame.display.set_mode((400,400))
screen.fill((0,0,0))
pygame.display.flip()
ciano = (000,255,255)
magenta = (255,000,255)
def dda(x1,y1,x2,y2):
    x,y = x1,y1
    length = (x2-x1)
    if length <= (y2-y1):
        length = y2-y1
    dx = (x2-x1)/float(length)
    dy = (y2-y1)/float(length)
    screen.set_at((round(x),round(y)),ciano)
    for i in range(length):
        x += dx
        y += dy
        screen.set_at((round(x),round(y)),ciano)
    pygame.display.flip()
def bresenham(x1,y1,x2,y2):
    screen.set_at((x1,y1),magenta)
    dx = x2-x1 #variação na direção x
    dy = y2-y1 #variação na direção y
    dy2 = 2*dy #dobro da variação de y 
    dx2 = 2*dx #dobro da variação de x
    dydx2 = dy2 - 2*dx #subtrai do dobro de y o dobro de x
    dxdy2 = dx2 - 2*dy #subtrai do dobro de x o dobro de y
    pox = dy2 - dx #po é igual a duas vezes a variação de y menos a variação de x 
    poy = dx2 - dy #po é igual a duas vezes a variação de x menos a variação de y
    x = x1
    y = y1
    if dx > dy:
        for i in range(dx):
            if pox < 0:
                screen.set_at((x+1,y),magenta)
                pox = pox + dy2
            else:
                screen.set_at((x+1,y+1),magenta)
                pox = pox + dydx2
                y += 1;
            x +=1;
        pygame.display.flip()
    else:
        for i in range(dy):
            if poy < 0:
                screen.set_at((x,y+1),magenta)
                poy = poy + dx2
            else:
                screen.set_at((x+1,y+1),magenta)
                poy = poy + dxdy2
                x += 1;
            y +=1;
        pygame.display.flip()
for i in range(80):
    bresenham(0,0,400,5*i)

for j in range(80):
    bresenham(0,0,5*j,400)

while 1:
    for event in pygame.event.get():
        if event.type == pygame.QUIT: sys.exit()
```
